# Task 4: Create separate array of Minimas and maximas

#### 4a Identify Minima and Maxima:



*   Create separate arrays to store the x-values of minima and maxima of the  function.
*   Print these arrays to display the critical points of the function.

```python
x = symbols('x')
from scipy.misc import derivative


# Define the function
def f(x):
    return x**5 - 6*x**3 + 9*x


# first diff
first_derivative = sp.diff(f(x), x)


# Second diff function
second_derivative = diff(first_derivative, x)




#Find turing point solve for dy/dx == 0

solutions = sp.solve(first_derivative, x)

# minimas and maximas

arr_minimas = []
arr_maximas = []

# minima and maxima points
for val in solutions:
    value = second_derivative.subs(x, val)
    if value > 0:
        arr_minimas.append((val, f(x).subs(x, val)))
    elif value < 0:
        arr_maximas.append((val, f(x).subs(x, val)))


print(f'the array of minimas {arr_minimas} and the array of maximas is {arr_maximas}')

## Jordan Nguepi
```

#### 4b Compute Global Minima and Maxima:



*   Determine the global minima and maxima of the quadratic function.
*   Print the coordinates of these points to identify the overall minimum and maximum values of the function.

```python
Global_maxima =  max(arr_maximas, key=lambda point: point[1])

Global_minima =  min(arr_minimas, key=lambda point: point[1])
print(f'The global maxima is {Global_maxima}') # (x, y)

print(f'The global minima is {Global_minima}') # (x, y)

# Jordan Nguepi
```
